<h1>Question 1</h1>

```python
import numpy as np
from statistics import mean, median, stdev

# Create a list of the first 20 natural numbers
numbers = list(range(1, 21))

# Sum of all even numbers
even_sum = sum(num for num in numbers if num % 2 == 0)

# Product of all odd numbers
odd_product = np.prod([num for num in numbers if num % 2 != 0])

# Mean, Median, Standard Deviation
mean_val = mean(numbers)
median_val = median(numbers)
stdev_val = stdev(numbers)

print(f"Even Sum: {even_sum}, Odd Product: {odd_product}, Mean: {mean_val}, Median: {median_val}, Std Dev: {stdev_val}")
```

<h1>Question 3</h1>

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]

# Plot
fib_seq = fibonacci(10)
plt.plot(fib_seq, marker='o')
plt.title("Fibonacci Sequence")
plt.show()
```

<h1>Question 10</h1>

```python
from collections import Counter
import matplotlib.pyplot as plt

def char_frequency(s):
    s = s.lower()
    freq = Counter(s)
    plt.bar(freq.keys(), freq.values())
    plt.title("Character Frequency")
    plt.show()
    return freq

char_frequency("Example String")
```

<h1>Question 12</h1>

```python
def analyze_strings(strings):
    longest = max(strings, key=len)
    vowel_start = sum(1 for s in strings if s[0].lower() in 'aeiou')

    lengths = [len(s) for s in strings]
    plt.hist(lengths, bins=5)
    plt.title("String Length Distribution")
    plt.show()

    return longest, vowel_start

strings = ["apple", "orange", "banana", "kiwi"]
print(analyze_strings(strings))
```

<h1>Question 14</h1>

```python
def filter_and_sort(sentence):
    words = [word for word in sentence.split() if len(word) >= 4]
    words.sort()
    lengths = [len(word) for word in words]

    plt.bar(words, lengths)
    plt.title("Word Lengths")
    plt.show()

    return words

print(filter_and_sort("This is a simple test example"))
```

<h1>Question 15</h1>

```python
def find_anagrams(word, words):
    sorted_word = sorted(word)
    return [w for w in words if sorted(w) == sorted_word]

print(find_anagrams("listen", ["enlist", "google", "inlets", "banana"]))
```

<h1>Question 18</h1>

```python
from collections import Counter
import matplotlib.pyplot as plt

# File Example
with open("text.txt", "w") as file:
    file.write("This is an example file. This file has words.")

with open("text.txt", "r") as file:
    words = file.read().lower().split()

word_count = Counter(words)

# Most Frequent Word
most_frequent = word_count.most_common(1)

# Bar Chart
plt.bar(word_count.keys(), word_count.values())
plt.title("Word Frequencies")
plt.show()

print(f"Most Frequent Word: {most_frequent}")
```

<h1>Question 22</h1>

```python
def normalize(lst):
    min_val, max_val = min(lst), max(lst)
    return [(x - min_val) / (max_val - min_val) for x in lst]

nums = [10, 20, 30, 40, 50]
normalized = normalize(nums)

plt.plot(normalized, marker='o')
plt.title("Normalized Data")
plt.show()

print(f"Normalized List: {normalized}")
```
